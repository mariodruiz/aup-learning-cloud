# ROCm Cell Profile — GEMM smoke test

Use this notebook to verify **jupyterlab-rocm** Cell Profile.

**How to profile**
1. Select a GEMM cell below.
2. Click the floating **Cell Profile** button beside the active code cell.
3. PyTorch GPU cells are profiled with `torch.profiler` in the live kernel; results appear inline and in **ROCm GPU Monitor → Profiler → Cell Profile**.

You can also run cells normally first (Shift+Enter) — Cell Profile still works on PyTorch GPU cells.

## 1. GPU check

```python
import torch

print(f"torch.cuda.is_available() = {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"device = {torch.cuda.get_device_name(0)}")
    print(f"device_count = {torch.cuda.device_count()}")
else:
    raise RuntimeError("No ROCm GPU visible to PyTorch — fix the runtime before profiling.")
```

## 2. Small GEMM

Select this cell and click **Cell Profile**.

```python
import torch

N = 1024
x = torch.randn(N, N, device="cuda", dtype=torch.float32)
y = x @ x
torch.cuda.synchronize()
print(f"GEMM {N}x{N} done, y[0,0] = {y[0, 0].item():.4f}")
```

## 3. Larger GEMM

```python
import torch

N = 2048
x = torch.randn(N, N, device="cuda", dtype=torch.float32)
y = x @ x
torch.cuda.synchronize()
print(f"GEMM {N}x{N} done, y[0,0] = {y[0, 0].item():.4f}")
```

## 4. Repeated GEMM loop

```python
import torch

M = 1024
a = torch.randn(M, M, device="cuda", dtype=torch.float16)
b = torch.randn(M, M, device="cuda", dtype=torch.float16)
for _ in range(8):
    c = a @ b
torch.cuda.synchronize()
print(f"8x fp16 GEMM {M}x{M} done, c[0,0] = {c[0, 0].item():.4f}")
```
